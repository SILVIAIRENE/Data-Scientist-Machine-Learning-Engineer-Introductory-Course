```python
import time

THICKNESS = 0.00008
FOLD_COUNT = 43

# Método 1: Operador de potencia
start_power = time.time()
folded_thickness_power = THICKNESS * (2 ** FOLD_COUNT)
elapsed_time_power = time.time() - start_power
print("Tiempo (potencia): {}[s]".format(elapsed_time_power))

# Método 2: Bucle for
start_loop = time.time()
folded_thickness_loop = THICKNESS
for _ in range(FOLD_COUNT):
    folded_thickness_loop *= 2
elapsed_time_loop = time.time() - start_loop
print("Tiempo (bucle): {}[s]".format(elapsed_time_loop))
```

```python
import matplotlib.pyplot as plt

THICKNESS = 0.00008
FOLD_COUNT = 43

thickness_list = [THICKNESS]
folded_thickness = THICKNESS

for _ in range(FOLD_COUNT):
    folded_thickness *= 2
    thickness_list.append(folded_thickness)

# Gráfico personalizado 1: Línea roja más gruesa
plt.title("Grosor del papel doblado")
plt.xlabel("Número de dobleces")
plt.ylabel("Grosor [m]")
plt.plot(thickness_list, color='red', linewidth=3)
plt.show()

# Gráfico personalizado 2: Línea punteada, fuente de ejes grande
plt.title("Grosor del papel doblado")
plt.xlabel("Número de dobleces")
plt.ylabel("Grosor [m]")
plt.tick_params(labelsize=15)
plt.plot(thickness_list, linestyle='--')
plt.show()

# Gráfico personalizado 3: Línea verde, marcadores cuadrados
plt.title("Grosor del papel doblado")
plt.xlabel("Número de dobleces")
plt.ylabel("Grosor [m]")
plt.plot(thickness_list, color='green', marker='s')
plt.show()
```
