2.El problema del bollo de castaña
Hay un episodio famoso sobre 2 a la enésima potencia [1] en el que aparece una herramienta llamada “Byvine” en Manga Doraemon.

[1] Fujiko F. Fujio (1979) “Doraemon (17)” Shogakukan.

“Byevine” es un medicamento, y cuando se rocía con él, la sustancia se duplica cada 5 minutos. Aumento el número de bollos de castaña en el proceso, pero como no podía comerlo, usé un cohete y lo tiré al espacio.

Un tema de discusión frecuente aquí es cómo aumentan de espacio después los bollos de castaña. Se han esgrimido varios argumentos, pero calculemos simplemente cómo aumentaría el volumen de un bollo de castaña.



Problema] El día en que los bollos de castaña cubren el sistema solar
Si espolvorea un solo bollo de castaña con vid, ¿cuánto tiempo estará cubierto el sistema solar con bollo de castaña? Crea un programa para calcular esto.

```python
import matplotlib.pyplot as plt
import numpy as np
import time

def calcular_tiempo_crecimiento(volumen_inicial, volumen_objetivo, tiempo_duplicacion_minutos=5):
    """
    Calcula el tiempo necesario para que un volumen crezca y alcance un volumen objetivo,
    duplicándose a intervalos regulares.

    Args:
        volumen_inicial (float): El volumen con el que se inicia.
        volumen_objetivo (float): El volumen que se quiere alcanzar.
        tiempo_duplicacion_minutos (int): El tiempo en minutos que tarda en duplicarse.

    Returns:
        tuple: Una tupla con (tiempo_final_minutos, tiempo_final_horas, volumenes_por_tiempo)
    """
    volumen_actual = volumen_inicial
    tiempo_transcurrido = 0
    volumenes_por_tiempo = []
    tiempos = []

    # Bucle principal para simular el crecimiento exponencial
    print("Iniciando el cálculo... Esto puede tardar unos segundos.")
    start_time = time.time()

    while volumen_actual < volumen_objetivo:
        volumenes_por_tiempo.append(volumen_actual)
        tiempos.append(tiempo_transcurrido)

        volumen_actual *= 2
        tiempo_transcurrido += tiempo_duplicacion_minutos

    end_time = time.time()
    print(f"Cálculo completado en {end_time - start_time:.2f} segundos.")

    tiempo_final_minutos = tiempo_transcurrido
    tiempo_final_horas = tiempo_final_minutos / 60

    return tiempo_final_minutos, tiempo_final_horas, volumenes_por_tiempo, tiempos

def graficar_crecimiento(tiempos, volumenes, nombre_objeto="bollo de castaña", nombre_objetivo="sistema solar"):
    """
    Genera un gráfico del crecimiento exponencial del volumen.
    """
    plt.style.use('seaborn-v0_8-darkgrid')
    plt.figure(figsize=(12, 7))

    # Convertir volúmenes a escala logarítmica para una mejor visualización
    volumenes_log = np.log10(volumenes)

    plt.plot(tiempos, volumenes_log, 'o-', color='#e67e22', markersize=4, linewidth=2, label=f'Volumen del {nombre_objeto} (log10)')

    plt.title(f'Crecimiento exponencial del volumen de un {nombre_objeto}', fontsize=18, fontweight='bold')
    plt.xlabel('Tiempo (minutos)', fontsize=14)
    plt.ylabel('Volumen (log10 en m³)', fontsize=14)

    plt.text(tiempos[-1], volumenes_log[-1], f'Tiempo final: {tiempos[-1]} min',
             ha='right', va='bottom', fontsize=12, color='#2c3e50', fontweight='bold')

    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

# --- DATOS DE ENTRADA ---
# Volúmenes en metros cúbicos (m³)
VOLUMEN_BOLLO = 3.35e-5  # Estimación de un bollo de castaña
VOLUMEN_SISTEMA_SOLAR = 3.8e38  # Estimación del sistema solar
VOLUMEN_TOKYO_DOME = 1.24e6  # Volumen del Tokyo Dome

# --- EJECUTAR EL CÁLCULO PARA EL PROBLEMA DEL BOLLO DE CASTAÑA ---
tiempo_final_min_bollo, tiempo_final_horas_bollo, vol_bollo_historial, tiempo_bollo_historial = \
    calcular_tiempo_crecimiento(VOLUMEN_BOLLO, VOLUMEN_SISTEMA_SOLAR)

print(f"\n--- Resultado para el bollo de castaña y el sistema solar ---")
print(f"El sistema solar estará cubierto de bollos de castaña en aproximadamente:")
print(f"  -> {tiempo_final_min_bollo} minutos")
print(f"  -> {tiempo_final_horas_bollo:.2f} horas")

# Generar el gráfico
graficar_crecimiento(tiempo_bollo_historial, vol_bollo_historial, nombre_objeto="bollo de castaña", nombre_objetivo="sistema solar")

# --- EJEMPLO GENERAL: TOKYO DOME Y UNA PELOTA DE BÉISBOL ---
# Puedes adaptar esta sección para diferentes problemas
VOLUMEN_PELOTA_BASEBALL = 3.8e-4  # Estimación de una pelota de béisbol
tiempo_final_min_pelota, tiempo_final_horas_pelota, vol_pelota_historial, tiempo_pelota_historial = \
    calcular_tiempo_crecimiento(VOLUMEN_PELOTA_BASEBALL, VOLUMEN_TOKYO_DOME)

print(f"\n--- Ejemplo: pelota de béisbol y el Tokyo Dome ---")
print(f"El Tokyo Dome se llenará con pelotas de béisbol en aproximadamente:")
print(f"  -> {tiempo_final_min_pelota} minutos")
print(f"  -> {tiempo_final_horas_pelota:.2f} horas")

# Generar el gráfico
graficar_crecimiento(tiempo_pelota_historial, vol_pelota_historial, nombre_objeto="pelota de béisbol", nombre_objetivo="Tokyo Dome")
```
