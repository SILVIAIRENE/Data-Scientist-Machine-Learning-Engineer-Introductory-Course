```python


!pip install -q missingno seaborn

# 1) Imports y detección de csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from scipy import stats

# Mostrar gráficos inline
%matplotlib inline
sns.set_context("notebook")

# Detectar archivo CSV: preferencia por 'train.csv', si no existe tomar primer .csv
cwd_files = os.listdir()
csv_candidates = [f for f in cwd_files if f.lower().endswith('.csv')]
if not csv_candidates:
    raise FileNotFoundError("No se encontró ningún archivo .csv en el directorio actual. "
                            "Sube el archivo 'train.csv' al entorno de Colab.")
# Preferir train.csv si existe
csv_name = 'train.csv' if 'train.csv' in csv_candidates else csv_candidates[0]
print(f"Usando archivo: {csv_name}")

# 2) Cargar dataset
df = pd.read_csv(csv_name)
print("\n== Dimensiones del dataset ==")
print(df.shape)

# 3) Inspección inicial del dataset (Problem 2 & Problem 3)
print("\n== Primeras 5 filas ==")
display(df.head())

print("\n== Tipos de datos por columna ==")
display(pd.DataFrame(df.dtypes, columns=['dtype']).T)

print("\n== Descripción estadística (numéricas) ==")
display(df.describe().T)

# Confirmar variable objetivo
target_col = "SalePrice"
if target_col not in df.columns:
    raise KeyError(f"No se encontró la variable objetivo '{target_col}' en el dataset.")
print(f"\nLa variable objetivo es: {target_col}")

# 4) Valores faltantes (Problem 4)
missing_counts = df.isnull().sum().sort_values(ascending=False)
missing_ratio = (missing_counts / len(df)).sort_values(ascending=False)
missing_df = pd.DataFrame({
    'TotalMissing': missing_counts,
    'MissingRatio': missing_ratio
})
print("\n== Valores faltantes por columna (top 30) ==")
display(missing_df[missing_df.TotalMissing>0].head(30))

# Visualizar con missingno (mapa de calor y matrix)
print("\nMostrando visualizaciones de missingno...")
plt.figure(figsize=(10,4))
msno.bar(df, fontsize=8)
plt.show()

plt.figure(figsize=(10,4))
msno.matrix(df, fontsize=8)
plt.show()

plt.figure(figsize=(10,6))
msno.heatmap(df)
plt.show()

# Seguir la instrucción: eliminar columnas con 5 o más valores faltantes (>=5)
threshold_cols = 5
cols_to_drop = missing_counts[missing_counts >= threshold_cols].index.tolist()
print(f"\nColumnas con >= {threshold_cols} valores faltantes (se eliminarán):")
print(cols_to_drop)

df_reduced = df.drop(columns=cols_to_drop)
print("\nDimensiones después de eliminar columnas con muchos missing:")
print(df_reduced.shape)

# Después de quitar columnas, eliminar filas que todavía tengan missing (según enunciado)
rows_before = len(df_reduced)
df_reduced = df_reduced.dropna(axis=0)
rows_after = len(df_reduced)
print(f"Filas antes de dropna: {rows_before}; filas después: {rows_after}; filas eliminadas: {rows_before-rows_after}")

# 5) Distribución de la variable objetivo (Problem 5 & 6)
y = df[target_col].dropna()  # original (antes de dropna en df_reduced)
print("\n== Estadísticas de SalePrice (original) ==")
print(y.describe())

skew_orig = stats.skew(y)
kurt_orig = stats.kurtosis(y)  # kurtosis de Fisher (exceso)
print(f"\nAsimetría (skew) original: {skew_orig:.4f}")
print(f"Curtosis (exceso) original: {kurt_orig:.4f}")

# Plot distribución original
plt.figure(figsize=(10,4))
sns.histplot(y, kde=True)
plt.title("Distribución de SalePrice (original)")
plt.xlabel("SalePrice")
plt.show()

# Transformación logarítmica (log1p para evitar problemas con 0)
y_log = np.log1p(y)

skew_log = stats.skew(y_log)
kurt_log = stats.kurtosis(y_log)
print(f"\nAsimetría (skew) log-transformada: {skew_log:.4f}")
print(f"Curtosis (exceso) log-transformada: {kurt_log:.4f}")

# Plot distribución log-transformada
plt.figure(figsize=(10,4))
sns.histplot(y_log, kde=True)
plt.title("Distribución de SalePrice (log1p transformada)")
plt.xlabel("log1p(SalePrice)")
plt.show()

print("""
Interpretación rápida:
- Si la distribución original tiene skew positivo alto y curtosis elevada (colas pesadas),
  aplicar log suele acercarla a una forma más simétrica y cercana a una normal.
- Esto ayuda a muchos algoritmos que asumen normalidad o reducen la influencia de outliers.
""")

# 6) Correlaciones (Problem 7)
# Usar el dataframe original (df) pero solo con columnas numéricas para correlación
numeric_df = df.select_dtypes(include=[np.number]).copy()
print("\nColumnas numéricas usadas para correlación:", len(numeric_df.columns))

corrmat = numeric_df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corrmat, vmax=0.8, square=True, cmap='vlag')
plt.title("Matriz de correlación (numéricas)")
plt.show()

# Seleccionar las top 10 características más correlacionadas con SalePrice (en valor absoluto)
corr_with_target = corrmat[target_col].abs().sort_values(ascending=False)
top10 = corr_with_target.index[1:11].tolist()  # [1:11] para excluir SalePrice mismo
print("\nTop 10 features más correlacionadas con SalePrice (abs corr):")
for i, col in enumerate(top10,1):
    print(f"{i}. {col} -> corr = {corrmat.loc[col, target_col]:.4f}")

# Heatmap solo para esas 10 + SalePrice
selected = top10 + [target_col]
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df[selected].corr(), annot=True, fmt=".2f", cmap='vlag', square=True)
plt.title("Heatmap correlaciones entre top10 features y SalePrice")
plt.show()

# Resumen interpretativo (breve) — sobre lo que representan en base a nombres de columnas
print("\nBreve interpretación de las top10 (consulte la descripción de Kaggle 'data description' para definiciones completas):")
for col in top10:
    print(f"- {col}")

# Detectar combinaciones altamente correlacionadas (posible multicolinealidad)
pairs = []
for i in range(len(top10)):
    for j in range(i+1, len(top10)):
        a = top10[i]; b = top10[j]
        r = numeric_df[[a,b]].corr().iloc[0,1]
        if abs(r) >= 0.8:
            pairs.append((a,b,r))
if pairs:
    print("\nPares entre top10 con correlación alta (>= 0.8):")
    for a,b,r in pairs:
        print(f"{a} <> {b} : corr = {r:.3f}")
else:
    print("\nNo se detectaron pares entre las top10 con correlación >= 0.8.")

print("""
Consideraciones para regresión múltiple:
- Si hay pares altamente correlacionados, considerar:
  * eliminar una de las columnas,
  * aplicar PCA o crear una variable compuesta (por ejemplo sumar/ratio),
  * usar regularización (Ridge/Lasso) para mitigar multicolinealidad.
""")

# 7) Guardar un pequeño dataset reducido (opcional) para modelos posteriores
reduced_for_model = numeric_df[selected].dropna()
reduced_for_model.to_csv("house_prices_top10_plus_target.csv", index=False)
print("\nSe guardó 'house_prices_top10_plus_target.csv' con las columnas seleccionadas (listas para modelado).")

# 8) Resumen final (texto listo para reporte)
print("\n=== RESUMEN FINAL (para incluir en un informe) ===\n")
print(f"- Dataset original: {df.shape[0]} filas x {df.shape[1]} columnas.")
print(f"- Columnas eliminadas por tener >= {threshold_cols} missing: {len(cols_to_drop)}.")
print(f"- Filas eliminadas tras dropna (después de eliminar columnas): {rows_before-rows_after}.")
print(f"- Asimetría de SalePrice original: {skew_orig:.3f}; curtosis (exceso): {kurt_orig:.3f}.")
print(f"- Tras transformación log1p: asimetría={skew_log:.3f}, curtosis={kurt_log:.3f}.")
print(f"- Top 10 variables más correlacionadas con SalePrice: {top10}.")
print("- Se ha guardado un CSV con las 10 features + SalePrice para modelado rápido.\n")

print("Fin del EDA. Puedes continuar con feature engineering (tratamiento detallado de categóricas), "
      "creación de pipelines y modelado (Regresión lineal, RandomForest, XGBoost, validación cruzada, etc.).")
```
