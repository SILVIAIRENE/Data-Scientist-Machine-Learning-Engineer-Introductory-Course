```python
import numpy as np
import matplotlib.pyplot as plt

# Se fija la semilla del generador de nmeros aleatorios para reproducibilidad
np.random.seed(0)
# [Problema 1] Crear nmeros aleatorios (Distribucin normal bidimensional)

# La media de la distribucin
mean_1 = [-3, 0]
# La matriz de covarianza
cov_1 = [[1.0, 0.8], [0.8, 1.0]]

# Crea 500 nmeros aleatorios con la distribucin bidimensional especificada
data_0 = np.random.multivariate_normal(mean_1, cov_1, 500)
print("Forma de data_0:", data_0.shape)

# [Problema 2] Visualizacin por diagrama de dispersin
# Visualiza los datos creados en el Problema 1
plt.figure(figsize=(8, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], alpha=0.8)
plt.title("Diagrama de dispersin de los datos del Problema 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()

# [Problema 3] Visualizacin por histograma
# Crea un histograma para cada dimensin (x1 y x2)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Histograma para la dimensin x1 (primera columna)
ax1.hist(data_0[:, 0], bins=50, rwidth=0.9)
ax1.set_title("histogram of x1")
ax1.set_xlabel("x1")
ax1.set_ylabel("frequency")
ax1.set_xlim(-6, 3) # Alinea el eje horizontal

# Histograma para la dimensin x2 (segunda columna)
ax2.hist(data_0[:, 1], bins=50, rwidth=0.9)
ax2.set_title("histogram of x2")
ax2.set_xlabel("x2")
ax2.set_ylabel("frequency")
ax2.set_xlim(-6, 3) # Alinea el eje horizontal

plt.tight_layout() # Asegura que los ttulos y etiquetas no se superpongan
plt.show()

# [Problema 4] Adicin de datos y visualizacin
# Parmetros para el segundo conjunto de datos
mean_2 = [0, -3]
cov_2 = [[1.0, 0.8], [0.8, 1.0]]
# Crea 500 nuevos nmeros aleatorios para la segunda distribucin
data_1 = np.random.multivariate_normal(mean_2, cov_2, 500)

# Visualiza ambos conjuntos de datos en un solo diagrama de dispersin
plt.figure(figsize=(8, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], label='0', alpha=0.8, c='tab:blue')
plt.scatter(data_1[:, 0], data_1[:, 1], label='1', alpha=0.8, c='tab:orange')
plt.title("Diagrama de dispersin con dos conjuntos de datos")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()

# [Problema 5] Combinacin de datos
# Combina los datos de los dos conjuntos verticalmente
combined_data = np.concatenate([data_0, data_1], axis=0)
print("Forma de los datos combinados:", combined_data.shape)
# [Problema 6] Etiquetado
# Crea las etiquetas (0s para data_0, 1s para data_1)
labels_0 = np.zeros((500, 1))
labels_1 = np.ones((500, 1))

# Combina los datos con sus respectivas etiquetas
data_with_labels_0 = np.concatenate([data_0, labels_0], axis=1)
data_with_labels_1 = np.concatenate([data_1, labels_1], axis=1)

# Une los dos conjuntos de datos con etiquetas para crear un ndarray final (1000, 3)
final_dataset = np.concatenate([data_with_labels_0, data_with_labels_1], axis=0)

print("\nPrimeros 5 elementos del conjunto de datos final:")
print(final_dataset[:5])
print("\nltimos 5 elementos del conjunto de datos final:")
print(final_dataset[-5:])
print("\nForma del conjunto de datos final (1000, 3):", final_dataset.shape)
```
